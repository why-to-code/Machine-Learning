```python
import numpy as np
x_values = np.arange(1,11)
x_values
```

```python
y_values = x_values *2
y_values
```

```python
plt.plot(x_values,y_values)
plt.show()
```

```python
plt.plot(x_values,y_values,color="g",linestyle="-.",linewidth=3)
plt.xlabel(" x- label")
plt.ylabel( " Y-label")
plt.title(" X vs Y")
plt.grid("True")
plt.show()
```

```python
y1_value = x_values *3
y1_value
```

```python
plt.plot(x_values,y_values,color="g",linestyle="-.",linewidth=3)
plt.plot(x_values,y1_value,color="r",linestyle="-.",linewidth=3)
plt.xlabel(" x- label")
plt.ylabel( " Y-label")
plt.title(" X vs Y")
plt.grid("True")
plt.show()
```

# Bar graph

```python
names = ["Amit","Sam","Rohan"]
marks = [45,54,50]
```

```python
plt.bar(names,marks)
plt.xlabel("student Names")
plt.ylabel(" Marks")
plt.title("Students VS Marks")
plt.show()
```

```python
plt.barh(names,marks)
plt.xlabel("student Names")
plt.ylabel(" Marks")
plt.title("Students VS Marks")
plt.show()
```

# Scatter

```python
x1_values = np.random.randint(2,10,(1,11))
x1_values
```

```python
y1_values = np.random.randint(2,10,(1,11))
y1_values
```

```python
plt.scatter(x1_values,y1_values,marker="*",s=100,c="r")
plt.show()
```

# Histogram

# Pie chart

```python
fruits = ["Mango","Orange","Apple"]
qt = [45,34,56]
```

```python
plt.pie(qt,labels=fruits,autopct="%0.01f%%")
plt.show()
```
